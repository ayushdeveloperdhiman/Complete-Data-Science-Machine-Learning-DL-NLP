```python
num = int(input("Enter a number:"))
factorial = 1
if num > 0:
    while num!=0:
        factorial *= num
        num -= 1
    print(factorial)
elif num == 0:
    print(factorial)
else:
    print("Number is negative")
```

```python
num = int(input("Enter a number"))
sum = 0
while num > 0:
    rem = num%10
    num //= 10
    sum += rem
print(f"Sum of digits is {sum}")
```

```python
num = int(input("Enter a number"))

if num > 1:
    for i in range(2, int(num**0.5) + 1):
        if num%i == 0:
            print(f'{num} is not a prime number')
            break
    else:
        print(f'{num} is a prime number')
else:
    print(f'{num} is not a prime number')
```
