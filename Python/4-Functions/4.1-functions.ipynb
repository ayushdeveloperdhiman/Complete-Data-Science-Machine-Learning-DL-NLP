##### Introduction to Functions
Definition:

A function is a block of code that performs a specific task.
Functions help in organizing code, reusing code, and improving readability.



```python
## Syntax
# function_name(parameter1, parameter2, ..., parameterN)
def function_name(parameters):
    # Function body
    """Docstring"""
    return expression
```

```python
## Why Functions?
num = 24
if num % 2 == 0:
    print(f"{num} is even")
else:
    print(f"{num} is odd")
```

```python
def even_or_odd(num):
    """This function finds if a number is even or odd."""
    if num % 2 == 0:
        print(f"{num} is even")
    else:
        print(f"{num} is odd")
```

```python
## Call this function
even_or_odd(24)
```

```python
## Functions with multiple parameters

def add(a,b):
    """This function adds two numbers."""
    return a + b
result = add(2, 4)
print(result)
```

```python
## Default Parameters

def greet(name):
    print(f"Hello, {name} Welcome to the paradise!")
greet("Ayush")
```

```python
def greet(name="Guest"):
    print(f"Hello, {name} Welcome to the paradise!")
greet("Ayush")
```

```python
## Variable length Arguments
## Positional and Keyword Arguments

def print_numbers(*ayush):
    for numbers in ayush:
        print(numbers)
```

```python
print_numbers(1,2,3,4,5,6,7,8,"Ayush")
```

```python
## Positional Arguments
def print_numbers(*args):
    for numbers in args:
        print(numbers)
```

```python
print_numbers(1,2,3,4,5,6,7,8,"Ayush")
```

```python
## Keyword Arguments

def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key} : {value}")
```

```python
print_details(name="Ayush", age=24, country="India")
```

```python
def print_details(*args,**kwargs):
    for val in args:
        print(f"Positional argument: {val}")

    for key,value in kwargs.items():
        print(f"{key} : {value}")
```

```python
print_details(1,2,3,4,"Ayush",name="Ayush", age=24, country="India")
```

```python
## Return statements
def multiply(a, b):
    return a * b

multiply(2, 4)
```

```python
## Return multiple values
def multiply(a, b):
    return a * b, a

multiply(2, 4)
```
