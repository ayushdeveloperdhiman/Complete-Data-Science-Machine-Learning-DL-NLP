#### Sets
Sets are a built-in data type in Python used to store collections of unique items. They are unordered, meaning that the elements do not follow a specific order, and they do not allow duplicate elements. Sets are useful for membership tests, eliminating duplicate entries, and performing mathematical set operations like union, intersection, difference, and symmetric difference.

```python
## create a set
my_set= {1, 2, 3, 4, 5}
print(my_set)
print(type(my_set))
```

```python
my_empty_set= set()
print(type(my_empty_set))
```

```python
my_set= set([1,2,3,4,5,6])
print(my_set)
```

```python
my_empty_set= set([1,2,3,6,5,4,5,6])
print(my_empty_set)
```

```python
## Basic set operations
## Adding and removing elements from a set
my_set.add(7)
print(my_set)
my_set.add(7)
print(my_set)  # No change, as sets do not allow duplicates
```

```python
## Removing the element
my_set.remove(3)
print(my_set)
```

```python
my_set.remove(10)
```

```python
my_set.discard(11)
print(my_set)
```

```python
## pop method
removed_element= my_set.pop()
print("Removed element:", removed_element)
print("Set after pop:", my_set)
```

```python
## clear all elements from the set
my_set.clear()
print("Set after clear:", my_set)
```

```python
## Set membership test
my_set= {1,2,3,4,5}
print(3 in my_set)  # True
print(10 in my_set)  # False
```

```python
## Mathematical set operations
set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9}

# Union
union_set = set1.union(set2)
print("Union:", union_set)

## Intersection
intersection_set = set1.intersection(set2)
print("Intersection:", intersection_set)

set1.intersection_update(set2)
print("Set1 after intersection_update:", set1)
```

```python
set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9}

## Difference
difference_set = set1.difference(set2)
print("Difference (set1 - set2):", difference_set)
```

```python
## Symmetric Difference
symmetric_diff_set = set1.symmetric_difference(set2)
print("Symmetric Difference:", symmetric_diff_set)
```

```python
## Sets methods

set1 = {1,2,3,4,5}
set2 = {3,4,5}

## is subset
print(set1.issubset(set2))

print(set1.issuperset(set2))
```

```python
lst = [1,2,2,3,4,4,5]

set(lst)
```

```python
## Counting unique words in text
text = "In this tutorial we are discussing about sets"
words = text.split()

## Convert list of words to set
unique_words = set(words)
print("Unique words:", unique_words)
print("Number of unique words:", len(unique_words))
```
