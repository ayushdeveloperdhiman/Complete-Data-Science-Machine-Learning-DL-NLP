##### Introduction to Dictionaries

Dictionaries are unordered collections of items. They store data in key-value pairs.
Keys must be unique and immutable (e.g., strings, numbers, or tuples), while values can be of any type.

```python
## Creating Dictionaries
empty_dict = {}
print(type(empty_dict))  # Output: <class 'dict'>
```

```python
empty_dict = dict()
empty_dict
```

```python
student = {"name": "Ayush", "age": 27, "grade": 24}
print(student)
print(type(student))  # Output: <class 'dict'>
```

```python
# Single key with multiple values (last one will overwrite)
student = {"name": "Ayush", "age": 27, "name": 24}
print(student)
```

```python
## Accessing Dictionary Elements
student = {"name": "Ayush", "age": 27, "grade": 'A'}
print(student)
```

```python
## Accessing values using keys
print(student["grade"]) # Accessing value using key 'grade'
print(student['age'])  # Accessing value using key 'age'

## Using get() method to access values
print(student.get("grade"))  # Accessing value using get() method for key 'grade'
print(student.get("last_name"))
print(student.get("last_name", "Not Available"))  # Providing default value if key doesn't exist
```

```python
## Modifying Dictionary Elements
## Dictonaries are mutable, so we can change their values
print(student)
```

```python
student['age'] = 33 ## Updating age from 27 to 33
print(student)
student["address"] = "india" ## Adding new key-value pair
print(student)
```

```python
del student['grade']  ## Deleting key-value pair with key 'grade'
print(student)
```

```python
## Dictionary Methods

keys = student.keys()
print(keys)  # Output: dict_keys(['name', 'age', 'address'])

values = student.values()
print(values)  # Output: dict_values(['Ayush', 33, 'india'])

items = student.items()
print(items)  # Output: dict_items([('name', 'Ayush'), ('age', 33), ('address', 'india')])
```

```python
## Shallow Copy
student_copy = student
print(student)
print(student_copy)
```

```python
student['name'] = "Ayush2"
print(student)
print(student_copy)
```

```python
student_copy1 = student.copy() ## Using copy() method to create a shallow copy
print(student_copy1)
print(student)
```

```python
student["name"] = "Ayush3"
print(student)
print(student_copy1)
```

```python
## Iterating over Dictionaries
## You can use loops to iterate over dictionaries keys, values, or items(key-value pairs).

## Iterating over keys
for keys in student.keys():
    print(keys)
```

```python
## Iterate over values
for value in student.values():
    print(value)
```

```python
## Iterate over key-value pairs

for key, value in student.items():
    print(f"{key}:{value}")
```

```python
## Netsed Dictionaries
students = {
    "student1": {"name": "Ayush", "age": 27, "grade": 'A'},
    "student2": {"name": "John", "age": 22, "grade": 'B'},
    "student3": {"name": "Alice", "age": 24, "grade": 'A+'}
}

print(students)
```

```python
## Accessing nested dictionary elements
print(students["student2"]["name"])  # Output: John
print(students["student2"]["age"])  # Output: 22
print(students["student2"]["grade"])  # Output: B
```

```python
students.items()
```

```python
## Iterating over nested dictionaries
for student_id, student_info in students.items():
    print(f"{student_id}: {student_info}")
    for key, value in student_info.items():
        print(f"{key}:{value}")  
    
```

```python
## Dictionary Comprehensions
squares = {x:x**2 for x in range(5)}
print(squares)
```

```python
## Conditional Dictionary Comprehensions
even_squares = {x:x**2 for x in range(10) if x%2 == 0}
print(even_squares)
```

```python
## Practical Examples

## Use a dictionary to count the frequency of elements in a list

numbers  = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
frequency = {}
for number in numbers:
    if number in frequency:
        frequency[number] += 1
    else:
        frequency[number] = 1
print(frequency)  # Output: {1: 1, 2: 2, 3: 3, 4: 4}
```

```python
## Merge 2 dictionaries into 1
dic1 = {"a": 1, "b": 2}
dic2 = {"b": 3, "c": 4}
merged_dict = {**dic1, **dic2}
print(merged_dict)
```
