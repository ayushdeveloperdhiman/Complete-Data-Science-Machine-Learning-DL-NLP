##### Introduction to Tuples
Explanation:

Tuples are ordered collections of items that are immutable.
They are similar to lists, but their immutability makes them different.


```python
## creating a tuple

empty_tuple = ()
print("Empty Tuple:", empty_tuple)
print("Type of empty_tuple:", type(empty_tuple))
```

```python
lst = list()
print(type(lst))
tpl = tuple()
print(type(tpl))
```

```python
numbers = tuple([1,2,3,4,5,6])
numbers
```

```python
list((1,2,3,4,5,6))
```

```python
mixed_tuple = (1, "Hello", 3.14, True)
print("Mixed Tuple:", mixed_tuple)
```

```python
## Accessing Tuple Elements
numbers
```

```python
print(numbers[2])
print(numbers[-1])
```

```python
numbers[0:4]
```

```python
numbers[::-1]
```

```python
## tuples operations

concatenation = numbers + mixed_tuple
print("Concatenation:", concatenation)
```

```python
mixed_tuple * 3
```

```python
numbers * 3
```

```python
## Immutable nature of Tuples
## Tuples are immutable, meaning once created, their elements cannot be changed, added, or removed.

lst = [1,2,3,4,5]
print(lst)

lst[1] = "Ayush"
print(lst)
```

```python
numbers[1] = "Ayush"
```

```python
## Tuple methods
print(numbers.count(1))
print(numbers.index(3))
```

```python
## Packing and Unpacking Tuples
##packing
packed_tuple = 1, "Hello", 3.14
print("Packed Tuple:", packed_tuple)
```

```python
## unpacking
a, b, c = packed_tuple
print(a)
print(b)
print(c)
```

```python
## Unpacking with Asterisk (*) Operator
numbers = (1, 2, 3, 4, 5, 6)
first, *middle, last = numbers
print("First:", first)
print("Middle:", middle)
print("Last:", last)
```

```python
## Nested Tuples
## Nested list
lst = [[1,2,3,4], [6,7,8,9], [1,"Hello",3.14,c]]
lst[0][0:3]
```

```python
lst = [[1,2,3,4], [6,7,8,9], (1,"Hello",3.14,c)]
lst[2][0:3]
```

```python
nested_tuple = ((1,2,3), ("a","b","c"), (True, False))

## Accessing elements in Nested Tuples
print(nested_tuple[0])        # First inner tuple
print(nested_tuple[1][2])     # Third element of second inner tuple
```

```python
## Iterating over nested tuples
for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item, end = " ")
    print()
```
