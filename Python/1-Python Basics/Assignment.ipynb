**Question 5:** Write a Python program to calculate the factorial of a number.

```python
# Your code here

def factorail(n):
    if n == 0:
        return 1
    else:
        return n * factorail(n-1)
num = int(input("Enter a number to calculate a factorial of a number"))
print(f"The factorial of {num} is {factorail(num)}")
```

**Question 17:** Write a Python program to check if a year is a leap year.

```python
# Your code here
year = int(input("Enter a year"))

if (year%4 == 0 and year %100 == 0) or (year%400 == 0):
    print(f'{year} is a leap year') 
else:
    print(f'{year} is not a leap year') 
```

**Question 18:** Write a Python program to reverse a string.

**Question 19:** Write a Python program to check if a string is a palindrome.

```python
# Your code here

str = input("Enter a string value")
if str.lower() == str[::-1].lower():
    print(f'{str} is a palindrome String')
else:
    print(f'{str} is not a palindrome string')
```
