Syntax refers to the set of rules that defines the combinations of symbols that are considered to be correctly structured programs in a
language. In simpler terms, syntax is about the correct arrangement of words and symbols in a code.

```python
## Basic Syntax Rules In Python
## Case sensitivity - Python is case sensitive

name="Ayush"
Name="Dhiman"

print(name)
print(Name)
```

```python
## Indentation
## Indentation is very important in Python. It is used to define the blocks of code.

age = 32

if age > 30:
    print(age)

print(age)
```

```python
## This is single line comment
print("Hello World")
```

```python
## Line continuation
## You can use backslash (\) to continue a statement on the next line.
total = 1 + 2 + 3 + 4 + 5 + 6 + 7 +\
4+5+6

print(total)
```

```python
## Multiple statements in a single line
x = 5; y = 10; z = x + y
print(z)
```
